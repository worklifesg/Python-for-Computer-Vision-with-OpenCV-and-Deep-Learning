### Object Detection with OpenCV and Python

In this section, we will cover the <b> Contour Detection</b>

- Contours are defined as <b> a curve joining all the continuous points </b> having same color or intensity
- Useful tool for shape analysis and object detection and recognition

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
img =cv2.imread('internal_external.png',0)
img.shape
```

```
plt.imshow(img,cmap='gray')
```

Here we can detect external contours (background) like triangles, circles and pie and internal contours (foreground) like smiley, pieces on pie.

```
contours,hierarchy = cv2.findContours(img,cv2.RETR_CCOMP,method=cv2.CHAIN_APPROX_SIMPLE)

# RETR_CCOMP - extract both external and internal contours

type(contours)
```

```
len(contours)
```

```
type(hierarchy)
```

```
hierarchy
```

```
external_contours = np.zeros(img.shape) #pure black

#to draw contours we should know how to read and imterpret contours and hierarchy

for i in range(len(contours)):

  if hierarchy[0][i][3] == -1: 
    cv2.drawContours(external_contours,contours,i,255,-1)

plt.imshow(external_contours,cmap='gray')
```

```
internal_contours = np.zeros(img.shape) #pure black

#to draw contours we should know how to read and interpret contours and hierarchy

for i in range(len(contours)):

  if hierarchy[0][i][3] != -1: 
    cv2.drawContours(internal_contours,contours,i,255,-1)

plt.imshow(internal_contours,cmap='gray')
```

<b> Please note:</b>

- <b> -1 </b> is for external contours (background)
- <b> Other than -1 </b> is for all internal contours
 - Else for specific internal contours, we need to check the hierarchy and run the code again as below

```
internal_contours_pie_pieces = np.zeros(img.shape) #pure black

#to draw contours we should know how to read and interpret contours and hierarchy

for i in range(len(contours)):

  if hierarchy[0][i][3] == 4: #4 assigned as for pie pieces
    cv2.drawContours(internal_contours_pie_pieces,contours,i,255,-1)

plt.imshow(internal_contours_pie_pieces,cmap='gray')
```

```
internal_contours_smile = np.zeros(img.shape) #pure black

#to draw contours we should know how to read and interpret contours and hierarchy

for i in range(len(contours)):

  if hierarchy[0][i][3] == 0: #0 assigned as for smile
    cv2.drawContours(internal_contours_smile,contours,i,255,-1)

plt.imshow(internal_contours_smile,cmap='gray')
```
